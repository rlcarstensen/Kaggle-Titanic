```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('./'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_data = pd.read_csv("./train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("./test.csv")
test_data.head()
```

```python
types = train_data.dtypes
print(types)

count_nan = len(train_data) - train_data.count()
print(count_nan)
```

```python
expanded_train_data = train_data
expanded_train_data = expanded_train_data.assign(has_age=expanded_train_data.Age.notna(),
                                                is_child=(expanded_train_data.Age.notna()) & (expanded_train_data['Age'] < 18),
                                                is_elderly=(expanded_train_data.Age.notna()) & (expanded_train_data.Age > 65),
                                                deck= expanded_train_data["Cabin"].str.split(r"(\d+| )",expand=True)[0].fillna('@').apply(lambda x: ord(x)-64),
                                                ticket_group = (expanded_train_data.Fare // 10 * 10).fillna(0.0))


expanded_train_data.head()
```

```python
expanded_test_data = test_data
expanded_test_data = expanded_test_data.assign(has_age=expanded_test_data.Age.notna(),
                                                is_child=(expanded_test_data.Age.notna()) & (expanded_test_data['Age'] < 18),
                                                is_elderly=(expanded_test_data.Age.notna()) & (expanded_test_data.Age > 65),
                                                deck=expanded_test_data["Cabin"].str.split(r"(\d+| )",expand=True)[0].fillna('@').apply(lambda x: ord(x)-64),
                                                ticket_group = (expanded_test_data.Fare // 10 * 10).fillna(0.0))
expanded_test_data.head()
```

```python
features = ["Pclass", "Sex", "SibSp", "Parch", "has_age", "deck", "ticket_group"]

print(pd.DataFrame(expanded_test_data, columns=features).isna().any())
print(pd.DataFrame(expanded_train_data, columns=features).isna().any())

y = expanded_train_data["Survived"]
X = pd.get_dummies(expanded_train_data[features])
X_test = pd.get_dummies(expanded_test_data[features])

def run_model_on_train(model):
    
    model.fit(X, y)
#     print(model)
    
    predictions = model.predict(X)
#     print(predictions)
    
    probability = model.predict_proba(X)
#     print(probability)
    
    score = model.score(X, y)
#     print(score)
    
    output = pd.DataFrame({'PassengerId': expanded_train_data.PassengerId, 'Survived': predictions})

    return model, predictions, probability, score, output


def run_model_on_test(model):
    
    model.fit(X, y)
#     print(model)
    
    predictions = model.predict(X_test)
#     print(predictions)
    
    probability = model.predict_proba(X_test)
#     print(probability)
    
    score = model.score(X, y)
#     print(score)
    
    output = pd.DataFrame({'PassengerId': expanded_test_data.PassengerId, 'Survived': predictions})

    return model, predictions, probability, score, output
```

```python
from sklearn.? import model_or_classifier

_model = model_or_classifier(random_state=0)

_results = run_model_on_train(_model)

_model = _results[0]
_predictions = _results[1]
_probability = _results[2]
_score = _results[3]
_output = _results[4]

print(_score)

_output
```

```python
from sklearn import tree
from sklearn.model_selection import cross_val_score

decision_tree_model = tree.DecisionTreeClassifier(random_state=0)

decision_tree_results = run_model_on_train(decision_tree_model)

decision_tree_model = decision_tree_results[0]
decision_tree_predictions = decision_tree_results[1]
decision_tree_probability = decision_tree_results[2]
decision_tree_score = decision_tree_results[3]
decision_tree_output = decision_tree_results[4]

print(decision_tree_score)

decision_tree_output
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forest_model = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=0)

random_forest_results = run_model_on_train(random_forest_model)

random_forest_model = random_forest_results[0]
random_forest_predictions = random_forest_results[1]
random_forest_probability = random_forest_results[2]
random_forest_score = random_forest_results[3]
random_forest_output = random_forest_results[4]

print(random_forest_score)

random_forest_output
```
